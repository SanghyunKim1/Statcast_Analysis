### 1.1 Data Type Conversion
Looking at the each data type, some of the nominal data attributes are treated as numerical data. I'm going to convert them from numerical data into categorical data, since nominal data are not comparable.
(see [here](https://baseballsavant.mlb.com/csv-docs)) <br />

```python
# convert some numerical data into nominal data
obj_cols = ['batter', 'pitcher', 'zone', 'hit_location', 'on_3b', 'on_2b', 'on_1b',
           'fielder_2', 'fielder_3', 'fielder_4', 'fielder_5', 'fielder_6', 'fielder_7',
            'fielder_8', 'fielder_9', 'sv_id', 'game_pk']

statcast_df[obj_cols] = statcast_df[obj_cols].astype('str')

print(statcast_df[obj_cols].dtypes)
```

### 1.2 Data Cleaning (Missing Data)
With more than 1,000,000 rows and 89 columns, such big (and raw) data are more likely to be imperfect.
Therefore, I'm going to first check how many missing data there are in our dataset.

```python
# total number of missing data
missing_cols = statcast_df.columns[statcast_df.isnull().any()]
print(statcast_df[missing_cols].isnull().sum())
```

As we see and expected, there seem to be many missing data in our dataset. However, it's not enough to just check the number of missing data. <br>
Thus, I'm going to visualize the missing data so I can get a more sense of them before we proceed to clean up data.

```python
# missing data visualization
msno.matrix(statcast_df[missing_cols], labels=True, fontsize=12, figsize=(25, 20), sort='descending')
plt.show()
```

```python
msno.bar(statcast_df[missing_cols], labels=True, fontsize=12, figsize=(25, 20))
plt.show()
```

By looking at those two graphs above, we can notice that the entire row is empty in specific columns (i.e. data attributes). <br>
In our dataset, <b>spin_dir</b>, <b>spin_rate_deprecated</b>, <b>break_angle_deprecated</b>, <b>break_length_deprecated</b>,<b>tfs_deprecated</b>, <b>tfs_zulu_deprecated</b> and <b>umpire</b> are such a case. As it's too dangerous to keep this column and to impute these missing data based on other data attributes, I'm going to delete these columns.

Also, there's one more thing worth paying attention. These two charts tell us the relationship betwen missing data. <br>
Look at the right hand side of the chart above. Apart from the attributes with the missingness proportion of 100% Data attributes located on the right hand side of the chart have simmilar numbers of missing values (the levels of missingness are balanced). For example, <b>launch_speed</b> and <b>launch_angle</b> have almost identical number of missing values indicating when there's a missing value in one column, then the other column is also missing. Same thing happens to <b>release_spin_rate</b> and <b>release_extension</b> and so on... <br>
Thus, it means that those missing data are unlikely to occur randomly. That is, such a type of misisng data is called <b>Missing Not At Random (MNAR)</b>. Let's have a look at a couple of more graphs to see if it's true. 

```python
# drop some data attributes with missing data proportion of 100%
cols_to_drop = [col for col in statcast_df[missing_cols].columns 
                if statcast_df[col].isnull().sum() == len(statcast_df[col])]
print(cols_to_drop)

statcast_df.drop(cols_to_drop, axis=1, inplace=True)
print(statcast_df.shape)
```

As we've dropped all the data attributes whose proportion of missing data is 100%, I'm going to check whether there's any relationship between missing data.

```python
# nullity correlation
missing_cols = statcast_df.columns[statcast_df.isnull().any()]
msno.heatmap(statcast_df[missing_cols], labels=True, fontsize=12, figsize=(25, 20))
plt.show()
```

The heatmap above indicates the <b>nullity correlation</b> between missing data attributes. Just like normal correlation coefficients, it varies from <b>-1</b> to <b>1</b>. <br>
As the correlation tends to -1, it indicates that if one variable is present, the other variable isn't (and vice versa). By contrast, the correlation of 1 shows that if one variable exists, then the other variable is also present. Having that said, the correlation closed to <b>0</b> shows that there's no <b> nullity relationship</b> between variables. <br>
By looking at the heatmap above, we can notice that some of the data attributes have high correlations in terms of <b>missingness</b> (e.g launch_speed and lauch_angle), while the missingness of some data fields are independent of the other variables.

To see clearer patterns between those missing values, let's see another diagram called <b>dendrogram</b>.

```python
# dendrogram
msno.dendrogram(statcast_df[missing_cols], fontsize=12, figsize=(25, 20))
plt.show()
```
